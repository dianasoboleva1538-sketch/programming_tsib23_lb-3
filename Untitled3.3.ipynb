```python
# Курс: ЦИБ - 251
# ФИО: Соболева Диана Юрьевна
# Тема Обработка исключений и утверждения (Задание task_03_03_01)
"""
Ошибки (номера строк через пробел, данная строка — №2): 5 6 7 8 9 10 11 12 13
"""
def sum_digits_less_5(n):
    """Вернуть сумму цифр, меньших 5, для положительного целого числа `n`.
    Если таких цифр нет, вернуть 0.
    """
    c = 0
    while n > 0:
        digit = n % 10
        if digit < 5:
            c += digit
        n //= 10
    return c
if __name__ == "__main__":
    print(sum_digits_less_5(1234))
    print(sum_digits_less_5(5678))
    print(sum_digits_less_5(104))
```
