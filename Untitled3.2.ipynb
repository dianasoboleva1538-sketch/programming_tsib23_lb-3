```python
# Курс: ЦИБ - 251
# ФИО: Соболева Диана Юрьевна
# Тема: сравнение сложности алгоритмов
# Алгоритм: подсчёт суммы всех цифровых символов в строке
# Сложность: O(N)
def sum_digits(s):
    val = 0
    for c in s:
        if c.isdigit():
            val += int(c)
    return val
print(sum_digits("a1b2c3"))
print(sum_digits("abc123def"))
print(sum_digits("hello world"))
print(sum_digits("456"))
print(sum_digits(""))
```
