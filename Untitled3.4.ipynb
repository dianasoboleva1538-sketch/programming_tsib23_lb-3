```python
# ФИО: Соболева Диана Юрьевна
# Группа: ЦИБ - 251
# Условие задачи: Измените задачу № 3.4.2 так, чтобы вне зависимости от ошибок чтения файла, программа выполняла подсчет суммы и максимума
# Задание task_03_04_03.
# Выполнил: Иванов И.О.
# Группа: ИТ-123

import os
def load_data(filename):
    """
    Загрузить список вещественных чисел из файла 'filename'.
 Сложность: O(N), где N — количество строк в файле.
    """
    numbers = []
    if not os.path.exists(filename):
        print(f"Файл '{filename}' не найден. Начинаем с пустого списка.")
        return numbers
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line_num, line in enumerate(file, 1):
                line = line.strip()

                # Пропускаем пустые строки
                if not line:
                    continue

                # Специальная обработка символа "-" как -1.0
                if line == '-':
                    numbers.append(-1.0)
                else:
                    try:
                        num = float(line)
                        numbers.append(num)
                    except ValueError:
                        # Не число и не "-" — пропускаем
                        pass

    except IOError as e:
        print(f"Ошибка ввода-вывода при чтении файла: {e}")
    except Exception as e:
        print(f"Непредвиденная ошибка при чтении: {e}")
    return numbers
def append_to_file(values, filename):
    """
    Дописать данные в конец файла.
    Параметры:
    - values (list): список значений для записи.
    - filename (str): имя файла.
    """
    try:
        with open(filename, 'a', encoding='utf-8') as file:
            for value in values:
                # Если значение целое число, записываем без десятичной точки
                if isinstance(value, float) and value.is_integer():
                    file.write(f"{int(value)}\n")
                else:
                    file.write(f"{value}\n")
    except IOError as e:
        print(f"Ошибка ввода-вывода при записи в файл: {e}")
    except Exception as e:
        print(f"Непредвиденная ошибка при записи: {e}")
def main():
    filename = input("Введите имя файла: ")
    numbers = load_data(filename)
    if numbers:
        total = sum(numbers)
        max_val = max(numbers)
        total_rounded = int(round(total))
    else:
        total = 0.0
        total_rounded = 0
        max_val = 0.0
    print(f"\nОбработка файла '{filename}' завершена.")
    print(f"  - Прочитано чисел: {len(numbers)}")
    if numbers:
        print(f"  - Список чисел: {numbers}")
    print(f"  - Сумма: {total} (округлённо: {total_rounded})")
    print(f"  - Максимум: {max_val}")
    print(f"  - Результаты дописаны в файл.")
    append_to_file([total_rounded, max_val], filename)
if __name__ == "__main__":
    main()
```
